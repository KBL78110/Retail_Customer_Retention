Botanic est un réseau de jardineries spécialistes du jardinage écologique. Présente en France avec plus de 75 magasins, Botanic propose des conseils en jardinage, des solutions naturelles et des produits écologiques pour le jardin. Ses cinq univers incluent le jardin et le potager, l’alimentation bio, les soins et le bien-être, les plantes et la décoration, ainsi que l’animalerie.

La construction d’un modèle prédictif d’attrition clients, telle que nous l'avons effectué, est un outil stratégique pour Botanic, lui permettant de maintenir sa base de clients fidèles tout en optimisant ses ressources.

L’objectif principal d’un tel modèle est d’anticiper le départ des clients avant qu’ils ne quittent effectivement l’enseigne. Cela permet à Botanic de prendre des mesures proactives pour retenir ces clients et maintenir leur fidélité.

```python
from datetime import datetime
# Convertir la colonne 'DATE_COMMANDE' en type datetime
df_BOTANIC_1['TIC_DATE'] = pd.to_datetime(df_BOTANIC_1['TIC_DATE'],dayfirst=True, errors='coerce')
# Filtrer les transactions effectuées en 2016
df_2016 = df_BOTANIC_1[(df_BOTANIC_1['TIC_DATE'] >= '2016-01-01') & (df_BOTANIC_1['TIC_DATE'] < '2017-01-01')]

# Filtrer les transactions effectuées en 2017 (si nécessaire)
df_2017 = df_BOTANIC_1[(df_BOTANIC_1['TIC_DATE'] >= '2017-01-01') & (df_BOTANIC_1['TIC_DATE'] < '2018-01-01')]
```

```python
# Calcul des Xi en 2016
Botanic_features_2016 = df_2016.groupby('IDCLIENT').agg(
    NBRE_COMMANDE_2016=pd.NamedAgg(column='TIC_DATE', aggfunc='count'),
    TOTAL_ACHAT_2016=pd.NamedAgg(column='TIC_TOTALTTC', aggfunc='sum'),
    PREMIERE_COMMANDE_2016=pd.NamedAgg(column='TIC_DATE', aggfunc='min'),
    DERNIERE_COMMANDE_2016=pd.NamedAgg(column='TIC_DATE', aggfunc='max')
).reset_index()
```

Calcul de l'ancienneté de la première et dernière commande pour chaque client à partir de la date de référence

```python
# Définir la date de fin 2016
end_date_2016 = pd.to_datetime('2016-12-31')

# 3. Calculer ANCIENNETE_1ERE_CDE et ANCIENNETE_DERNIERE_CDE
Botanic_features_2016['ANCIENNETE_1ERE_CDE_2016'] = (end_date_2016 - Botanic_features_2016['PREMIERE_COMMANDE_2016']).dt.days
Botanic_features_2016['ANCIENNETE_DERNIERE_CDE_2016'] = (end_date_2016 - Botanic_features_2016['DERNIERE_COMMANDE_2016']).dt.days

# 4. Afficher le DataFrame mis à jour
print(Botanic_features_2016)
```

Calcul du temps moyen entre 2 commandes en 2016

```python
#Calcul du temps moyen entre 2 commandes en 2016
Botanic_features_2016['TEMPS_MOY_CDE_2016']= (Botanic_features_2016['ANCIENNETE_1ERE_CDE_2016']- Botanic_features_2016['ANCIENNETE_DERNIERE_CDE_2016']) /(Botanic_features_2016['NBRE_COMMANDE_2016']-1)
# Affichage du DataFrame final
print(Botanic_features_2016)
```

```python
import pandas as pd

# Vérifiez s'il y a des zéros dans 'ANCIENNETE_DERNIERE_CDE_2016'
nb_zeros_anciennete = (Botanic_features_2016['ANCIENNETE_DERNIERE_CDE_2016'] == 0).sum()
print(f"Nombre de zéros dans 'ANCIENNETE_DERNIERE_CDE_2016': {nb_zeros_anciennete}")

# Vérifiez les types de données
print("Types de données:")
print(Botanic_features_2016[['TEMPS_MOY_CDE_2016', 'ANCIENNETE_DERNIERE_CDE_2016']].dtypes)

# Si nécessaire, corrigez les types de données
Botanic_features_2016['TEMPS_MOY_CDE_2016'] = Botanic_features_2016['TEMPS_MOY_CDE_2016'].astype(float)
Botanic_features_2016['ANCIENNETE_DERNIERE_CDE_2016'] = Botanic_features_2016['ANCIENNETE_DERNIERE_CDE_2016'].astype(float)

# Remplacez les zéros dans 'ANCIENNETE_DERNIERE_CDE_2016' par une petite valeur (ex: 1)
Botanic_features_2016.replace({'ANCIENNETE_DERNIERE_CDE_2016': {0: 1}}, inplace=True)
```
