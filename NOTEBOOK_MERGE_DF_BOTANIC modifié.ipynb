Botanic est un réseau de jardineries spécialistes du jardinage écologique. Ses cinq univers incluent le jardin et le potager, l’alimentation bio, les soins et le bien-être, les plantes et la décoration, ainsi que l’animalerie.

La construction d’un modèle prédictif d’attrition clients, telle que nous l'avons effectué, est un outil stratégique pour Botanic, lui permettant de maintenir sa base de clients fidèles tout en optimisant ses ressources.

L’objectif principal d’un tel modèle est d’anticiper le départ des clients avant qu’ils ne quittent effectivement l’enseigne. Cela permet à Botanic de prendre des mesures proactives pour retenir ces clients et maintenir leur fidélité.

Transposition de la colonne "CODEUNIVERS" pour faciliter la suite des jointures

```python
#Transposition de la colonne "CODEUNIVERS" pour faciliter la suite des jointures
df_LIGNE_ART["codeunivers_U1"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U1',1,0)
df_LIGNE_ART["codeunivers_U2"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U2',1,0)
df_LIGNE_ART["codeunivers_U3"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U3',1,0)
df_LIGNE_ART["codeunivers_U4"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U4',1,0)
```

Aggrégation de df_CLIENT_MAG_ENT_LIG_ART par "IDTICKET"

```python
#Aggrégation de df_CLIENT_MAG_ENT_LIG_ART par "IDTICKET"
df_LIGNE_ART=df_LIGNE_ART.groupby(["IDTICKET"]).agg(
    Quantite=('QUANTITE','sum'),
    NBRE_ARTICLE=('IDARTICLE', 'nunique'),
    NBRE_FAM_ARTICLE=('CODEFAMILLE','nunique'),
    UNIVERS_U1=('codeunivers_U1','sum'),
    UNIVERS_U2=('codeunivers_U2','sum'),
    UNIVERS_U3=('codeunivers_U3','sum'),
    UNIVERS_U4=('codeunivers_U4','sum'),
    TOTAL_REMISE=('MONTANTREMISE','sum'),
    MONTANT_TOTAL=('TOTAL','sum'),
    MARGE=('MARGESORTIE','sum')
).reset_index()
df_LIGNE_ART.head()
```

JOINTURE DES DATAFRAMES

```python
# Appliquer melt pour les univers
df_LIGNE_ART_melted = df_LIGNE_ART.melt(
    id_vars=['IDTICKET'],  
    value_vars=['UNIVERS_U1', 'UNIVERS_U2', 'UNIVERS_U3', 'UNIVERS_U4'],
    var_name='TYPE_UNIVERS',
    value_name='CODEUNIVERS'
)

# Affichage pour vérification
print(df_LIGNE_ART_melted)

# Joints le résultat fondu sur les autres colonnes d'intérêt
df_LIGNE_ART_melted = df_LIGNE_ART_melted.join(df_LIGNE_ART.set_index('IDTICKET'), on='IDTICKET', lsuffix='_left', rsuffix='_right')

# Enlever les colonnes univers éclatées initiales après le merge
df_LIGNE_ART_melted = df_LIGNE_ART_melted.drop(['UNIVERS_U1', 'UNIVERS_U2', 'UNIVERS_U3', 'UNIVERS_U4'], axis=1)

# Afficher le DataFrame final pour vérification
print(df_LIGNE_ART_melted)
```

```python

df_2017['TIC_DATE'] = pd.to_datetime(df_2017['TIC_DATE'], dayfirst=True)

# Calculer le nombre de commandes en 2017 pour chaque client
Botanic_nbre_cde_2017 = df_2017.groupby('IDCLIENT').size().reset_index(name='NBRE_CDE_2017')
print(Botanic_nbre_cde_2017)
```

```python
# Convertir la colonne de date en datetime
df_2017['DATEREADHESION'] = pd.to_datetime(df_2017['DATEREADHESION'], dayfirst=True)

# Filtrer pour garder uniquement les commandes de 2017
df_2017 = df_2017[df_2017['DATEREADHESION'].dt.year == 2017]

# Fixer la date de référence pour 2017 (fin de 2017)
end_of_2017 = pd.Timestamp('2017-12-31')

# Calculer l'ancienneté en 2017 pour chaque client par rapport à DATEREADHESION
df_2017['ANCIENNETE_READ_2017'] = (end_of_2017 - df_2017['DATEREADHESION']).dt.days
# Convertir les jours en mois (~30.44 jours par mois)
df_2017['ANCIENNETE_READ_2017_MONTHS'] = (df_2017['ANCIENNETE_READ_2017'] / 30.44).astype(int)
# Prendre la première date d'adhésion pour chaque client
client_anciennete = df_2017.groupby('IDCLIENT')['ANCIENNETE_READ_2017'].first().reset_index()
# Fusionner avec le DataFrame contenant le nombre de commandes
result = pd.merge(Botanic_nbre_cde_2017, client_anciennete, on='IDCLIENT')

print(result)
```

```python
# Convertir les jours en mois (~30.44 jours par mois)
df_2017['ANCIENNETE_READ_2017_MONTHS'] = (df_2017['ANCIENNETE_READ_2017'] / 30.44).astype(int)

# Prendre la première ancienneté en mois pour chaque client
client_anciennete = df_2017.groupby('IDCLIENT')['ANCIENNETE_READ_2017_MONTHS'].first().reset_index()

# Fusionner avec le DataFrame contenant le nombre de commandes
result = pd.merge(Botanic_nbre_cde_2017, client_anciennete, on='IDCLIENT')

# Discrétiser l'ancienneté en mois en catégories
bins = [0, 2, 6, 10, float('inf')]
labels = ['0-2 mois', '2-6 mois', '6-10 mois', 'plus de 10 mois']
result['ANCIENNETE_READ_CAT_2017'] = pd.cut(result['ANCIENNETE_READ_2017_MONTHS'], bins=bins, labels=labels, right=False)
```
