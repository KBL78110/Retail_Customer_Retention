Transposition de la colonne "CODEUNIVERS" pour faciliter la suite des jointures

```python
#Transposition de la colonne "CODEUNIVERS" pour faciliter la suite des jointures
df_LIGNE_ART["codeunivers_U1"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U1',1,0)
df_LIGNE_ART["codeunivers_U2"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U2',1,0)
df_LIGNE_ART["codeunivers_U3"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U3',1,0)
df_LIGNE_ART["codeunivers_U4"]=np.where(df_LIGNE_ART["CODEUNIVERS"]=='U4',1,0)
```

Aggrégation de df_CLIENT_MAG_ENT_LIG_ART par "IDTICKET"

```python
#Aggrégation de df_CLIENT_MAG_ENT_LIG_ART par "IDTICKET"
df_LIGNE_ART=df_LIGNE_ART.groupby(["IDTICKET"]).agg(
    Quantite=('QUANTITE','sum'),
    NBRE_ARTICLE=('IDARTICLE', 'nunique'),
    NBRE_FAM_ARTICLE=('CODEFAMILLE','nunique'),
    UNIVERS_U1=('codeunivers_U1','sum'),
    UNIVERS_U2=('codeunivers_U2','sum'),
    UNIVERS_U3=('codeunivers_U3','sum'),
    UNIVERS_U4=('codeunivers_U4','sum'),
    TOTAL_REMISE=('MONTANTREMISE','sum'),
    MONTANT_TOTAL=('TOTAL','sum'),
    MARGE=('MARGESORTIE','sum')
).reset_index()
df_LIGNE_ART.head()
```

JOINTURE DES DATAFRAMES

```python
# Appliquer melt pour les univers
df_LIGNE_ART_melted = df_LIGNE_ART.melt(
    id_vars=['IDTICKET'],  
    value_vars=['UNIVERS_U1', 'UNIVERS_U2', 'UNIVERS_U3', 'UNIVERS_U4'],
    var_name='TYPE_UNIVERS',
    value_name='CODEUNIVERS'
)

# Affichage pour vérification
print(df_LIGNE_ART_melted)

# Joints le résultat fondu sur les autres colonnes d'intérêt
df_LIGNE_ART_melted = df_LIGNE_ART_melted.join(df_LIGNE_ART.set_index('IDTICKET'), on='IDTICKET', lsuffix='_left', rsuffix='_right')

# Enlever les colonnes univers éclatées initiales après le merge
df_LIGNE_ART_melted = df_LIGNE_ART_melted.drop(['UNIVERS_U1', 'UNIVERS_U2', 'UNIVERS_U3', 'UNIVERS_U4'], axis=1)

# Afficher le DataFrame final pour vérification
print(df_LIGNE_ART_melted)
```

FILTRAGE DES COMMANDES DE 2016

```python
from datetime import datetime
# Convertir la colonne 'DATE_COMMANDE' en type datetime
df_BOTANIC_1['TIC_DATE'] = pd.to_datetime(df_BOTANIC_1['TIC_DATE'],dayfirst=True, errors='coerce')
# Filtrer les commandes effectuées en 2016
df_2016 = df_BOTANIC_1.loc[(df_BOTANIC_1['TIC_DATE'] >= '2016-01-01') & (df_BOTANIC_1['TIC_DATE'] < '2017-01-01')]

# Affichage du DataFrame filtré
print(df_2016)
```

Calcul de l'ancienneté de la première commande pour chaque client à partir de la date de référence

```python

# Conversion de la colonne TIC_DATE en format datetime avec dayfirst=True
# Utilisation de .loc pour éviter l'avertissement SettingWithCopyWarning
df_2016.loc[:, 'TIC_DATE'] = pd.to_datetime(df_2016['TIC_DATE'], dayfirst=True)

# Définir la date de référence comme le 31/12/2016
date_reference = pd.to_datetime('31/12/2016', dayfirst=True, errors="coerce")
# Calculer l'ancienneté de la première commande pour chaque client à partir de la date de référence
anciennete_premiere_commande = df_2016.groupby('IDCLIENT')['TIC_DATE'].min().reset_index()
anciennete_premiere_commande['ANCIENNETE_1ERE_COMMANDE'] = (
    date_reference - anciennete_premiere_commande['TIC_DATE']
).dt.days 

# Vérifier que les colonnes sont correctes
print(anciennete_premiere_commande.columns)  
# S'assurer que le DataFrame contient les bonnes colonnes
DATE_MIN = anciennete_premiere_commande[['IDCLIENT', 'ANCIENNETE_1ERE_COMMANDE']]

# Merge des données sur l'IDCLIENT pour ajouter colonne ANCIENNETE_1ERE_COMMANDE aux commandes de 2016
df_BOTANIC_1 = df_BOTANIC_1.merge(DATE_MIN, on='IDCLIENT', how='right')

# Affichage du DataFrame final
print(df_BOTANIC_1)
```

Calcul de l’ancienneté de la dernière commande pour chaque client à partir de la date de référence

```python
# Conversion de la colonne TIC_DATE en format datetime avec dayfirst=True
# Utilisation de .loc pour éviter l'avertissement SettingWithCopyWarning
df_2016.loc[:, 'TIC_DATE'] = pd.to_datetime(df_2016['TIC_DATE'], dayfirst=True)
# Définir la date de référence comme le 31/12/2016
date_reference = pd.to_datetime('31/12/2016', dayfirst=True, errors="coerce")
# Calculer l’ancienneté de la dernière commande pour chaque client à partir de la date de référence
anciennete_dernière_commande = df_2016.groupby('IDCLIENT')['TIC_DATE'].max().reset_index()
anciennete_dernière_commande['ANCIENNETE_DERNIERE_COMMANDE'] = (
    date_reference - anciennete_dernière_commande['TIC_DATE']
).dt.days 

# Vérifier que les colonnes sont correctes
print(anciennete_dernière_commande.columns)  
# S'assurer que le DataFrame contient les bonnes colonnes
DATE_MAX = anciennete_dernière_commande[['IDCLIENT', 'ANCIENNETE_DERNIERE_COMMANDE']]
# Merge des données sur l'IDCLIENT pour ajouter colonne ANCIENNETE_DERNIERE_COMMANDE aux commandes de 2016
df_BOTANIC_1 = df_BOTANIC_1.merge(DATE_MAX, on='IDCLIENT', how='right')

# Affichage du DataFrame final
print(df_BOTANIC_1)

```

Calcul nombre de commande pour chaque client à partir de la date de référence

```python
import pandas as pd

# Vérification 'TIC_DATE' est de type datetime dans les deux DataFrames
df_2016['TIC_DATE'] = pd.to_datetime(df_2016['TIC_DATE'], dayfirst=True)
df_BOTANIC_1['TIC_DATE'] = pd.to_datetime(df_BOTANIC_1['TIC_DATE'], dayfirst=True)

# Groupement par IDCLIENT et TIC_DATE
order_counts = df_2016.groupby(['IDCLIENT', 'TIC_DATE']).size().reset_index(name='NBRE_CDE')

# Fusion des DataFrames sans duplication des colonnes de fusion
df_BOTANIC_1 = df_BOTANIC_1.merge(order_counts, on=['IDCLIENT', 'TIC_DATE'], how='left')

# Affichage du DataFrame mis à jour pour vérifier
print(df_BOTANIC_1.head())
```

Calcul du temps moyen entre 2 commandes

```python
df_BOTANIC_1['TEMPS_MOY_CDE']= (df_BOTANIC_1['ANCIENNETE_1ERE_COMMANDE']- df_BOTANIC_1['ANCIENNETE_DERNIERE_COMMANDE']) /(df_BOTANIC_1['NBRE_CDE']-1)
# Affichage du DataFrame final
print(df_BOTANIC_1)
```

CALCUL DE LA FRÉQUENCE DES COMMANDES

```python
# Création de la variable 'FREQ_COMMANDE' basée sur la colonne 'TEMPS_MOY_CDE'
df_BOTANIC_1['FREQ_COMMANDE'] = df_BOTANIC_1['TEMPS_MOY_CDE']/df_BOTANIC_1['ANCIENNETE_DERNIERE_COMMANDE']

# Afficher le DataFrame avec la nouvelle variable 'COMMANDE'
df_BOTANIC_1['FREQ_COMMANDE']
```

Calcul de la variete_panier pour chaque client

```python
# Calculer le nombre unique de catégories de produits pour chaque client
variete_panier = df_BOTANIC_1.groupby('IDCLIENT')['NBRE_ARTICLE'].nunique().reset_index()

# Renommer la colonne résultante à quelque chose de descriptif
variete_panier.columns = ['IDCLIENT', 'Variete_Panier']

print("Variété du panier pour chaque client")
print(variete_panier)
# Effectuer un merge pour intégrer variete_panier au DataFrame original
df_BOTANIC_1 = df_BOTANIC_1.merge(variete_panier, on='IDCLIENT', how='right')

# Affichage du DataFrame final
print(df_BOTANIC_1)
```

Calcul du nombre d'univers distincts dans lesquels chaque client a acheté

```python
# Calcul du nombre d'univers distincts dans lesquels chaque client a acheté
univers_par_client = df_BOTANIC_1.groupby('IDCLIENT')['CODEUNIVERS'].nunique().reset_index()
univers_par_client.columns = ['IDCLIENT', 'Nb_Univers']

print("Nombre d'univers distincts pour chaque client")
print(univers_par_client)
# Effectuer un merge pour intégrer Nb_Univers au DataFrame original
df_BOTANIC_1 = df_BOTANIC_1.merge(univers_par_client, on='IDCLIENT', how='right')

# Affichage du DataFrame final
print(df_BOTANIC_1)
```

Calcul de l'univers préféré pour chaque client

```python
# Grouper par client et univers, et sommer le nombre d'articles
client_univers = df_BOTANIC_1.groupby(['IDCLIENT', 'CODEUNIVERS'])['NBRE_ARTICLE'].sum().reset_index()

# Trouver l'univers préféré (celui avec le maximum de NBRE_ARTICLE) pour chaque client
univers_prefere = client_univers.loc[client_univers.groupby('IDCLIENT')['NBRE_ARTICLE'].idxmax()]

# Renommer les colonnes pour plus de clarté
univers_prefere.columns = ['IDCLIENT', 'Univers_Prefere', 'Total_Articles']

print("Univers préféré pour chaque client")
print(univers_prefere)
# Effectuer un merge pour intégrer univers_prefere au DataFrame original
df_BOTANIC_1 = df_BOTANIC_1.merge(univers_prefere, on='IDCLIENT', how='right')

# Affichage du DataFrame final
print(df_BOTANIC_1)
```
